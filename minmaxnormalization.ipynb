```python
import pandas as pd
```

```python
df = pd.read_csv("Academic_Performance.csv")
```

```python
df
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Gender'].fillna('Unknown',inplace=True)
df['Address'].fillna('Unknown',inplace=True)
df['Division'].fillna('Unknown',inplace=True)
```

```python
df['English'].fillna(df['English'].mean(),inplace=True)
df['Science'].fillna(df['Science'].mean(),inplace=True)
```

```python
#Find outliers uisng zscore method
from scipy.stats import zscore

z_scores = zscore(df['Math'])
outliers = df[abs(z_scores) > 1]
print(outliers)
```

```python
#Find outliers using Inter Quartile Range (IQR)
Q1 = df['Math'].quantile(0.25)
Q3 = df['Math'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Math'] < lower_bound) | (df['Math'] > upper_bound)]
print(outliers)
```

```python
#Min-max data normalization

normalized_math = (df['Math'] - df['Math'].min()) / (df['Math'].max() - df['Math'].min())

df['Math'] = normalized_math
print(df)
```
